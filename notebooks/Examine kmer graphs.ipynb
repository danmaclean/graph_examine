```python
import igraph as ig
import structs
import pickle
import os
import pandas as pd
import copy as cp
import itertools
```

## Data loading

1. Make a list of filenames (data structures are precalculated and pickled using the HPC and structs).
2. Make an `igraph` object from each file.
3. Extract subgraphs of kmers with coverage $>1$



```python
def make_igraph(kmer_info):
    '''makes an igraph from a kmer overlap object'''
    g = ig.Graph(directed=True)
    kmers = kmer_info.kmer_attrs.keys()
    id_map = {kmers[i] : i for i in range(len(kmers)) }
    g.add_vertices(len(kmers))
    g.vs["origin"] = origin_list(kmers, kmer_info)
    g.vs["coverage"] = get_coverage(kmers, kmer_info)
    g["origins"] = unique_origin_list(kmers,kmer_info)
    g.add_edges([(id_map[i[0]],id_map[i[1]]) for i in kmer_info.edges ])
    return g

def reduce_graph(graph, edge_count):
    g = cp.deepcopy(graph)
    v = [v.index for v in g.vs.select(lambda v: v.degree() == edge_count)]
    g.delete_vertices(v)
    return g

def origin_list(kmers, kmer_info):
    '''returns list of sequence origins for all vertices'''
    return [kmer_info.kmer_attrs[str(kmer)]['origin'].keys() for kmer in kmers]

def unique_origin_list(kmers,kmer_info):
    '''returns list of unique origin names'''
    l = origin_list(kmers,kmer_info)
    return list(set(itertools.chain(*l)))

def get_coverage(kmers, kmer_info):
    return [kmer_info.kmer_attrs[str(kmer)]['coverage'] for kmer in kmers]

files = [f for f in os.listdir(".") if "pickled_" in f ]
graph_count = len(files)
graphs = [make_igraph(pickle.load(open(f,'rb'))) for f in files ]
reduced_graphs = [reduce_graph(g, 2) for g in graphs]

```

## Build dataframe of graph statistics

```python
def build_dataframe(files):
    df = pd.DataFrame()
    df['genes_used'] = [f.split("_")[1] for f in files]
    df['iterations'] = [f.split("_")[2] for f in files]
    df['initial_nodes'] = [g.vcount() for g in graphs]
    df['initial_edges'] = [g.ecount() for g in graphs]
    df['multi_nodes'] = [g.vcount() for g in reduced_graphs]
    df['multi_edges'] = [g.ecount() for g in reduced_graphs]
    df['initial_assortativity'] = [g.assortativity_degree() for g in graphs]
    df['multi_assortativity'] = [g.assortativity_degree() for g in reduced_graphs]
    df['max_degree'] = [g.maxdegree() for g in graphs]
    df['multi_max_degree'] = [g.maxdegree() for g in reduced_graphs]
    #df['degree_distribution'] = [g.degree_distribution() for g in graphs]
    df['initial_components'] = [len(g.clusters().sizes() ) for g in graphs]
    df['multi_components'] = [len(g.clusters().sizes() ) for g in reduced_graphs]
    df['modularity'] = [g.clusters().q  for g in graphs]
    df['multi_modularity'] = [g.clusters().q  for g in reduced_graphs]
    #df['component_distribution'] = [g.clusters().sizes() for g in graphs]
    pickle.dump( df, open( "dataframe.pkl", "wb" ) )
    return df

def reload_from_pickle(pickle_file):
    return pickle.load(open(pickle_file,'rb'))

try:
    df = reload_from_pickle("dataframe.pkl")
except:
    df = build_dataframe(files)
    
print df.to_string()
```

## Some methods for getting data out of dataframe

```python
def get_levels(df,column,sort=True):
    '''gets the levels for a factor in 'column' '''
    levels = list(set(df[column]))
    if sort:
        return sorted(levels)
    else:
        return levels

def split_data_on(df, column,sort=True):
    '''splits the dataframe into subframes based on the name of 'column' '''
    levels = get_levels(df,column,sort)
    return [df[getattr(df,column) == l] for l in levels]
    
def get_column_data(df_list,column):
    '''gets data from  'column' for each dataframe element in the list'''
    return [list(df[column]) for df in df_list]



```
