# **_R_** gene assembly as a metagenomics problem


**_R_** genes are plant resistance genes. Plants have hundreds of these things. It is a feature of **_R_** genes in that they are structurally very similar (particularly at the self-similar C-terminal leucine-rich repeat), so there are a lot of shared k-mers between different genes, so they are hard to assemble.

In metagenomics, a central problem is to isolate short(-ish) sequencing reads in a mixed sample into groups whose members all come from a common (ideally single) organism.  In metagenomics sequence assembly difficulties arrive because some sequence reads or _k-mers_will ambiguously appear to come from different organisms. 

We can map this idea onto the **_R_** gene assembly concept by noting that within a given plant genome the **_R_** genes are equivalent to organisms and we are trying to separate k-mers of these reads back to their original gene of origin, given that we start off with an unsorted mess of sequences.

One practical approach to solving is to examine _k-mer_ graphs of **_R_** genes. 

The central question of this analysis is:

**_Can the constituent k-mers be partitioned back into the original genes in the same way that sequence reads from a metagenomics sample can be partitioned back into the original genes or species?_**


## Approach

The approach I will take then is to assume

1. that we will always be working from sequence reads with no known reference genome, 
2. that we will always have reads from some sort of enrichment experiment (so nearly all **_R_** gene fragments)
3. that we will always work on one species at a time

A popular and flexible system for this sort of metagenomics partitioning and analysis is _khmer_. 

[http://khmer.readthedocs.org/en/v1.1/](http://khmer.readthedocs.org/en/v1.1/)

## Functions for this notebook

I keep almost all functions here for ease of reference

```python
def get_hist_dataframe(file):
    """ returns a pandas dataframe from a khmer .hist file"""
    with open(file, "r") as f:
        df = pd.DataFrame()
        contents = [line.split() for line in f.readlines() ]
        df["occurs"] = [int(l[0]) for l in contents]
        df["times_observed"] = [int(l[1]) for l in contents]
        df["running_total"] = [int(l[2]) for l in contents]
        df["cumul_freq"] = [float(l[3]) for l in contents]
        f.close()
        return df

def write_generic_script(command):
    """writes a generic script that can be run using the %run line magic. The %run magic doesn't appear to allow
    string interpolation or command-line args so this function and script wraps that functionality so we can just call 
    one script with no args"""
    f = open("generic_script.py", "w")
    text = """
#!/usr/bin/env python
import subprocess
command = "{0}"
p = subprocess.Popen(command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
out, err = p.communicate()

o = open("stdout_dump","w")
if out:
    o.write(out)
o.close()

e = open("stderr_dump","w")
if e:
    e.write(str(err))
e.close()
""".format(command)
    f.write(text)
    f.close()

def print_file_contents(file):
    """prints a file's contents to a cell"""
    lines = open(file, "r").readlines()
    print "".join(lines)
    
def report_dump():
    """reports the results of the stderr and stdout dump to a cell.
    sends files to print_file_contents"""
    for f in ["stdout_dump","stderr_dump"]:
        print_file_contents(f)

def run_generic_script(command):
    """runs generic_script.py in the background, collects reports from STDOUT and STDERR and cleans up."""
    try:
        write_generic_script(command)
        %run generic_script
        report_dump()
    except:
        "External script didn't run correctly. You may have a rogue generic_script.py in cwd"
    finally:
        os.remove("generic_script.py")
        os.remove("stdout_dump")
        os.remove("stderr_dump")
        
def run_load_into_counting(input_file, k=31, n=4, min_tbl_size= 2e6,out="out.tabl"):
    """runs the khmer load into counting script"""
    command = "python {0} --ksize {1} --n_tables {2} --min-tablesize {3} {4} {5}".format(scripts["load_into_counting"], k, n, min_tbl_size, out, input_file)
    run_generic_script(command)
    
def run_abundance_dist(input_table, input_sequences, out_hist="out.hist"):
    """runs the khmer adundance dist script". Returns a pandas dataframe."""
    command = "python {0} {1} {2} {3}".format(scripts["abundance_dist"], input_table, input_sequences, out_hist)
    run_generic_script(command)
    return get_hist_dataframe(out_hist)

def run_normalize_by_median(input_file, k=31,n=4,min_tbl_size=2e6,c=20,out="out.kh"):
    """runs the khmer normalize by median script."""
    command = "python {0} --ksize {1} --n_tables {2} --min-tablesize {3} -C {4} -o {5} {6}".format(scripts["normalize_by_median"],k, n, min_tbl_size, c, out, input_file) 
    run_generic_script(command)

def run_do_partition(input_file, k=31,n=4,min_tbl_size=2e6,out="out.part"):
    """runs the khmer do partition script"""
    command = "python {0} --ksize {1} --n_tables {2} --min-tablesize {3} {4} {5}".format(scripts["do_partition"],k,n,min_tbl_size,out,input_file )
    run_generic_script(command)
    
def run_extract_partitions(input_file, max_size=1, min_size=1, basename="parts"):
    command = "python {0} --max-size {1} --min-partition-size {2} {3} {4}".format(scripts["extract_partitions"],max_size,min_size, basename, input_file)
    run_generic_script(command)


def run_load_graph(input_file, k=31,n=4,min_tbl_size=2e6, basename="base"):
    command = "python {0} --ksize {1} --n_tables {2} --min-tablesize {3} {4} {5}".format(scripts["load_graph"],k, n, min_tbl_size, basename, input_file)
    run_generic_script(command)
    
def run_make_initial_stoptags(basename="base"):
    command = "python {0} {1}".format(scripts["stop_tags"], basename)
    run_generic_script(command)
    
def run_partition_graph(basename="base"):
    command = "python {0} {1}".format(scripts["partition_graph"], basename)
    run_generic_script(command)

def run_find_knots(basename="base"):
    command = "python {0} {1}".format(scripts["find_knots"], basename)
    run_generic_script(command)
    
def run_filter_stoptags(input_file, k=31,basename="base"):
    tags = basename + ".stoptags"
    command = "python {0} --ksize {1} {2} {3}".format(scripts["filter_stoptags"],k, tags, input_file)
    run_generic_script(command)

def plot_hist(df):
    plt.plot(df.occurs, df.times_observed,label="kmer count frequency")
    plt.plot(df.occurs,  df.running_total,label="kmer count cumulative frequency")
    plt.ylabel('times observed')
    plt.xlabel('kmer occurs')
    plt.legend(loc=4)
    plt.show()
    
def make_annotation_dataframe(file):
    """Loads in the PRGDB annotation files and prepares a dataframe.
    Headers are PRGID\tName\tType\tSpecies\tClass\tGenBank ID\tGenBank Locus\tDescription
    """
    df = pd.DataFrame()
    f = open(file, "r")
    contents = [line.split('\t') for line in f.readlines() ]
    df['PRGID'] = [l[0] for l in contents]
    df['name'] = [l[1] for l in contents]
    df['type'] = [l[2] for l in contents]
    df['species'] = [l[3] for l in contents]
    df['class'] = [l[4] for l in contents]
    df['genbank_id'] = [l[5] for l in contents]
    df['genbank_locus'] = [l[6] for l in contents]
    df['description'] = [l[7] for l in contents]
    return df

def find_genome_project(organism):
    """looks up organism at Entrez genomes. Returns False if nothing found. Returns True if
    find a hit with a non-empty Accession ID.
    Returns None in all other cases."""
    try:
        handle = Entrez.esearch(db="genome", term=organism) 
        info = Entrez.read(handle)
        if 'WarningList' in info and 'No items found.' in info['WarningList']['OutputMessage']:
            return False
        else:
            handle2 = Entrez.esummary(db="genome", id=info['IdList'][0])
            record = Entrez.read(handle2)
            if record[0]["Assembly_Accession"] != "":
                return True
        return None
    except:
        return None

def is_a_plant(organism):
    """Takes an organism name and returns True if organism can be positively identified as a plant.
    False if positively not a plant.
    Returns None if something goes wrong with the look up"""
    ##find the organism taxid
    try:
        handle = Entrez.esearch(db="taxonomy", term=organism)
        info = Entrez.read(handle)
        handle = Entrez.efetch(db="taxonomy", id=info['IdList'][0])
        record = Entrez.read(handle)
        if "Viridiplantae" in record[0]["Lineage"]:
            return True
        else:
            return False
    except:
        return None

def select_from_fasta(input_fasta, gene_ids):
    """given a list of gene_ids, selects those from a fasta and writes just those back to a fasta file"""
    seq_list = list(SeqIO.parse(input_fasta, "fasta"))
    keepers = [i for i in seq_list if i.id in gene_ids]
    return keepers

def sequence_count_fasta(input_fasta):
    """given a fasta file, returns number of sequences in it"""
    return len(list(SeqIO.parse(input_fasta, "fasta")))

def write_fasta(seq_list, filename):
    """writes fasta file from a list of Bio.SeqIO objects"""
    output_handle = open(filename, "w")
    try:
        count = SeqIO.write(seq_list, output_handle, "fasta")
        output_handle.close()
    except IOError as e:
        print "I/O error({0}): {1}".format(e.errno, e.strerror)
        sys.exit
    return True

def count_partitions(in_file):
    prog = re.compile('^>(.*)\s+(\d+)\n')
    parts = {}
    seqs = []
    with open(in_file, "r") as f:
        for line in f.readlines():
            m = prog.match(line)
            if m:
                seqs.append(m.group(1))
                if int(m.group(2)) in parts:
                    parts[int(m.group(2))].append(m.group(1))
                else:
                    parts[int(m.group(2))] = [m.group(1)]
        return max(parts.keys()), list(set(seqs)), parts
        
def clean(file_list):
    for f in file_list:
        os.remove(f)
```

A. The genes are in just one partition anyway ... see below...

```python
def find_smallest_partition(partition_dict):
    genes = list(set([i for i in itertools.chain.from_iterable(partitions.values()) ]))
    smallest_partitions = []
    for g in genes:
        count = 0
        for k in partition_dict:
            if g in partition_dict[k]:
                count +=1
        smallest_partitions.append([g, count])
    return smallest_partitions

find_smallest_partition(partitions)
```
